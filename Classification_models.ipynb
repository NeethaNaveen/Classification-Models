```python
import seaborn as sns
import pandas as pd
```

# load data

```python
from sklearn.datasets import  load_breast_cancer
```

```python
data= load_breast_cancer()
```

```python
df=pd.DataFrame(data.data,columns=data.feature_names)
```

```python
df
```

```python
# Add the target variable to the DataFrame
df['target'] = data.target
df['target']
```

# assigning features as X and target as y

```python
X =df.drop('target', axis=1)
y=df['target']
```

# scaling feature data

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

# Splitting data 

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X_scaled,y,test_size=0.2, random_state=42)
```

#  Classification models


# Logistic Regression

 It is a linear model that predicts probabilities of belonging to a particular class. It's suitable for binary classification problems and is often used as a baseline model.


```python
from sklearn.linear_model import LogisticRegression
```

```python
log_reg=LogisticRegression()
log_reg.fit(X_train,y_train)
```

```python
log_pred=log_reg.predict(X_test)
log_pred
```

# DecisionTreeClassifier

It is a tree-based model that splits data into subsets based on feature values. It's suitable for datasets with a mix of categorical and numerical features.

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dtc=DecisionTreeClassifier()
dtc.fit(X_train,y_train)
```

```python
dtc_pred=dtc.predict(X_test)
dtc_pred
```

# RandomForestClassifier

It is an ensemble model that combines multiple Decision Trees. It's suitable for datasets with a large number of features and can handle missing values.


```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rfc=RandomForestClassifier()
rfc.fit(X_train,y_train)
```

```python
rfc_pred=rfc.predict(X_test)
rfc_pred
```

# Support Vector Classifier

It  is a linear or non-linear model that finds the hyperplane that maximally separates classes. It's suitable for datasets with a small number of features and can handle high-dimensional data.


```python
from sklearn.svm import SVC
```

```python
svc=SVC()
svc.fit(X_train,y_train)
```

```python
svc_pred=svc.predict(X_test)
svc_pred
```

# KNeighborsClassifier

It is a lazy model that predicts the class of a new instance based on the majority vote of its k nearest neighbors. It's suitable for datasets with a small number of features and can handle missing values.


```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn=KNeighborsClassifier(n_neighbors=5)
knn.fit(X_train,y_train)
```

```python
knn_pred=knn.predict(X_test)
knn_pred
```
